```python
import pandas as pd

dfu = pd.read_csv('2020-11-16-DC-URLs.csv')

dfu = dfu.dropna()

dfu
```

```python
import altair as alt

alt.Chart(dfu).mark_line().encode(
    alt.X('Time:T', title='Date'),
    alt.Y('finished:Q', title='URLs Downloaded')
).properties(
    width=500,
    height=300
)
```

```python
import pandas as pd

dfb = pd.read_csv('2020-11-16-DC-Bytes.csv')

dfb = dfb.dropna()

# Turn to TB:
dfb['tb'] = dfb['total'] / 1e12

dfb
```

```python
import altair as alt

alt.Chart(dfb).mark_line().encode(
    alt.X('Time:T', title='Date'),
    alt.Y('tb:Q', title='Bytes Downloaded (TB)')
).properties(
    width=500,
    height=300
)
```
